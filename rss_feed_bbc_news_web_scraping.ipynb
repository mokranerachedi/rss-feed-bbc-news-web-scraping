```python
# Website URL
url = "http://feeds.bbci.co.uk/news/rss.xml"
```

```python
# Import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get(url)
response
```

```python
# Creating a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# Getting list of item 
items = soup.findAll('item')
items
```

```python
data = []

for i in items:
    item = {
        'title': i.title.text,
        'description': i.description.text,
        'pub_date': i.pubdate.text,
        'link': i.guid.text
    }
    data.append(item)
```

```python
# Creating a dataframe from data
df = pd.DataFrame(data, columns=['title', 'description', 'pub_date', 'link'])
df
```

```python
# Export as excel file
df.to_excel('rss_feeds_bcc_news.xlsx')
```
